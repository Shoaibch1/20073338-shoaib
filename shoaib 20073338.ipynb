```python
# Import the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Agriculture land aera data set dataset
agr= pd.read_csv('/content/Agriculture.csv')
```

```python
#start from column 4
agr=agr.iloc[4:]
```

```python
#show the table
agr.head()
```

```python
#skips the 1st 4 columns of the table for year
agr_year=agr.iloc[:,4:]
```

```python
#show the table of year
agr_year.head()
```

```python
#tranpose the year columns
agr_year.T
```

```python
#finding the sum of year at zero axis
sum_of_agr_year=agr_year.sum(axis=0)
sum_of_agr_year
```

```python
#assign new variable to insert in to new col
agr_year_land=pd.DataFrame(sum_of_agr_year)
agr_year_land
```

```python
#drop the NAN fromt he table
agr_year_land=agr_year_land.dropna()
agr_year_land
```

```python
#change the data type of the  float to int
agr_year_land['year']=agr_year_land.index.astype('int')
agr_year_land
```

```python
#data type check
agr_year_land.dtypes
```

```python
#ploting
plt.figure(figsize=(8,5))
plt.plot(agr_year_land['year'],agr_year_land[0])
plt.title('Total Agriculture land per year')
plt.xlabel('Year')
plt.ylabel('Total Agriculture area')
plt.show()
```

```python
#now plotting the countries name code or 1998 year
agr_country_1998=agr[['Country Name','Country Code','1998']]
```

```python
agr_country_1998
```

```python
#selected the some countries
agr_country_code_1998=['CHN','USA','PAK','AFG','RUS','GBR']
```

```python
#country name and code with area land in 1998
agr_year_country_1998=agr_country_1998[agr_country_1998['Country Code'].isin(agr_country_code_1998)]
agr_year_country_1998
```

```python
#PLOTTING 1998 area land
fig, ax = plt.subplots(figsize=(9,4))
sns.barplot(x="Country Name", y="1998", data=agr_year_country_1998)
ax.set_title("Agriculture land(%) by Country name",fontdict= {'size':12})
ax.xaxis.set_label_text("Countries",fontdict= {'size':12})
ax.yaxis.set_label_text("Land Area(%) in 1998",fontdict= {'size':12})
plt.show()
```
